```python
from sklearn.decomposition import PCA  # For performing Principal Component Analysis

from mpl_toolkits.mplot3d import Axes3D  # For 3D plotting in Matplotlib

import pandas as pd  # For data wrangling and manipulation using DataFrames
import matplotlib.pyplot as plt  # For data visualisation with various plot types
```

# PCA

### BLOOD 


```python
# Fitting the PCA model to the data
pca.fit(df1)

# Printing the proportion of variance captured by each principal component
print('Proportion of variance captured by each PC:')
print(pca.explained_variance_ratio_)

# Plotting the proportion of variance explained by each principal component
plt.plot(range(1,11), pca.explained_variance_ratio_)
plt.xlabel('Principal component')
plt.ylabel('Explained variance')
plt.show()

# Defining a list of sample names
sample_names = df1.index
# Generating a score matrix by transforming the data using the fitted PCA model
df1_scores = pd.DataFrame(pca.transform((df1)), index=sample_names).T

# Creating a dictionary that maps each sample to its disease state
state = dict(zip(bl_df.columns, bl_df.loc["Sample_Type"]))
# Assigning colours to each sample based on its disease state
colour_key     = {'0': 'green', '1': 'red'}
# generating a list of sample colours using list comprehension
sample_colours = [colour_key[state[i]] for i in sample_names]

# Plotting a scatter plot of samples, where their position is determined by the first two principal component scores
plt.scatter(df1_scores.iloc[0], df1_scores.iloc[1], c=sample_colours)
# show the variance captured by each PC in the axis labels
plt.xlabel('PC1 ('+str(round(pca.explained_variance_ratio_[0]*100))+'%)')
plt.ylabel('PC2 ('+str(round(pca.explained_variance_ratio_[1]*100))+'%)')
plt.show()
```
